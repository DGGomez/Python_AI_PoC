Imports

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing
import datetime as dt
from sklearn import utils
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
scaler = MinMaxScaler(feature_range=(0, 1))
lab_enc = preprocessing.LabelEncoder()

plt.style.use('seaborn-whitegrid')

def selection_sort(x):
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
    return x
```

Load data

```python
# holder
dataX = []
dataY = []
dataZ = []

dates1 = []
dates2 = []

# load data and match data to dates
trainingData=open('data/data.txt', 'r') # You may wish to specify types, or process columns once read
for line in trainingData:
    split = line.split(" ")
    dataX.append(float(split[0]))
    dataY.append(float(split[1]))
    dates1.append(split[2][0:2]+"-"+split[2][2:4]+"-"+split[2][4:8])
    dataZ.append(float(split[2][4:8]+split[2][2:4]+split[2][0:2]))

trainingData.close()

# sort stuff

sortX = np.array(dataX)
sortX = selection_sort(sortX)

sortZ = np.array(dataZ)
sortZ = selection_sort(sortZ)
```

Initial Analysis

```python
f1 = plt.figure()
plt.title("Overlap of crime and price")
plt.plot(dataX, linewidth = 0.5)
plt.plot(dataY, linewidth = 0.1)

plt.show()

# daily changes
changeX = []
changeY = []

adjustedX = []
adjustedY = []
for i in range(len(dataX)-1):
    valX = dataX[i+1] - dataX[i]
    valY = dataY[i+1] - dataY[i]
    changeX.append(valX)
    changeY.append(valY)
    if(valX >0): 
        adjustedX.append(1)
    else:
        adjustedX.append(-1)
    if(valY >0): 
        adjustedY.append(1)
    else:
        adjustedY.append(-1)

f2 = plt.figure()
plt.title("Changes day-to-day (zoomed in)")
plt.plot(changeX[100:200], linewidth = 0.5)
plt.plot(changeY[100:200], linewidth = 0.5)

plt.show()

f3 = plt.figure()
plt.title("Changes day-to-day Price")
plt.plot(changeX, linewidth = 0.5)

plt.show()

f4 = plt.figure()
plt.title("Changes day-to-day Crime")
plt.plot(changeY, linewidth = 0.5)

plt.show()

f5 = plt.figure()
plt.title("Adjusted Changes day-to-day (zoomed in)")
plt.plot(adjustedX[40:50], linewidth = 0.5)
plt.plot(adjustedY[40:50], linewidth = 0.5)

plt.show()

f3 = plt.figure()
plt.title("Changes day-to-day Price")
plt.plot(adjustedX[100:200], linewidth = 0.5)

plt.show()

f4 = plt.figure()
plt.title("Changes day-to-day Crime")
plt.plot(adjustedY[100:200], linewidth = 0.5)

plt.show()
```

Crime VS Price

```python
# normalize data
# average crime vs price (sort by price)
trainingX = sortX
trainingY = [0]*len(dataY)

x = []
y = []
tx = []
ty = []
# times counted
z = [0]*len(trainingY)

for i in range(len(trainingX)):
    # find index in original set
    index = dataX.index(trainingX[i])
    if(dataY[index] in trainingY):
        inner = trainingY.index(dataY[index])
        trainingY[inner] += dataY[index]
        z[i] += 1
    else:
        trainingY[i] = dataY[index]
        z[i] = 1
    x.append([float(trainingX[i])])
    y.append([float(trainingY[i])/z[i]])

tx = x[3000:]
x = x[:3000]
ty = y[3000:]
y = y[:3000]

trainingX = np.array(x)
trainingY = np.array(y)
testingX = np.array(tx)
testingY = np.array(ty)

svr_lin = SVR(kernel='linear', C=33, gamma='auto')
svr_rbf = SVR(kernel='rbf', C=33, gamma=0.1, epsilon=.1)

X_scaled = preprocessing.scale(trainingX)
test_scaled = preprocessing.scale(testingX)

svr_lin.fit(X_scaled, trainingY)
svr_rbf.fit(X_scaled, trainingY)

predictionLin = svr_lin.predict(test_scaled)
predictionrbf = svr_rbf.predict(test_scaled)

# error
print("Linear")
print("Mean squared error: %.2f"
      % mean_squared_error(testingY, predictionLin))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(testingY, predictionLin))
print()
print("RBF")
print("Mean squared error: %.2f"
      % mean_squared_error(testingY, predictionrbf))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(testingY, predictionrbf))

# plot data
f1 = plt.figure()
plt.xlabel("Price")
plt.ylabel("Crime")

plt.title("Price vs Crime")
plt.scatter(test_scaled, testingY, color = 'black', label='Data')
plt.plot(test_scaled, predictionrbf, color = 'red', label='RBF model')
plt.plot(test_scaled, predictionLin, color = 'green', label='Linear model')
plt.show()
```

```python
# date vs price (sort by price)
df = pd.read_csv('spx.csv')
data = df.sort_index(ascending=True, axis=0)
new_data = pd.DataFrame(index=range(0,len(df)),columns=['Date', 'Close'])

svr_lin = SVR(kernel='linear', C=45, gamma='auto')
svr_poly = SVR(kernel='poly', C=45, degree=3, gamma= 'auto', epsilon=.1, coef0=1)
svr_rbf = SVR(kernel='rbf', C=45, gamma=0.1, epsilon=.1)

for i in range(0,len(data)):
    new_data['Date'][i] = data['Date'][i]
    new_data['Close'][i] = data['Close'][i]

plt.figure(figsize=(16,8))
plt.plot(df['Close'], label='Close Price history')

for i in range(len(new_data['Date'])):
    split = new_data['Date'][i].split("-")
    new_data['Date'][i] = float(split[0]+split[1]+split[2])
    
train = new_data[:20000]
subtrain = new_data[19000:20000]
valid = new_data[20000:]

x_train = train.drop('Close', axis=1)
subx_train = subtrain.drop('Close', axis=1)

y_train = train['Close']
suby_train = subtrain['Close']

x_test = valid.drop('Close', axis=1)
y_test = valid['Close']

X_scaled = preprocessing.scale(x_train)
test_scaled = preprocessing.scale(x_test)

reg = LinearRegression()
reg.fit(x_train,y_train)
preds = reg.predict(x_test)
    
valid['Predictions'] = 0
valid['Predictions'] = preds

valid.index = new_data[20000:].index
train.index = new_data[:20000].index

svr_lin.fit(X_scaled, y_train)
svr_poly.fit(X_scaled, y_train)
svr_rbf.fit(X_scaled, y_train)

predictionPoly = svr_poly.predict(test_scaled)
predictionLin = svr_lin.predict(test_scaled)
predictionrbf = svr_rbf.predict(test_scaled)

f1 = plt.figure()
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Linear Regression")

plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.show()

print("POLY")
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, predictionPoly))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, predictionPoly))
print()
print("Linear")
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, predictionLin))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, predictionLin))
print()
print("RBF")
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, predictionrbf))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, predictionrbf))

f2 = plt.figure()
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("SVR")

plt.plot(test_scaled, y_test)
plt.plot(test_scaled, predictionPoly, color = 'blue', label='Polynomial model')
plt.plot(test_scaled, predictionLin, color = 'green', label='Linear model')
plt.plot(test_scaled, predictionrbf, color = 'red', label='RBF model')
plt.show()

# remove lagging bit
X_scaled = preprocessing.scale(subx_train)
test_scaled = preprocessing.scale(x_test)

svr_lin.fit(X_scaled, suby_train)
svr_poly.fit(X_scaled, suby_train)
svr_rbf.fit(X_scaled, suby_train)

predictionPoly = svr_poly.predict(test_scaled)
predictionLin = svr_lin.predict(test_scaled)
predictionrbf = svr_rbf.predict(test_scaled)

print("POLY")
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, predictionPoly))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, predictionPoly))
print()
print("Linear")
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, predictionLin))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, predictionLin))
print()
print("RBF")
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, predictionrbf))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, predictionrbf))

f3 = plt.figure()
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Adjusted SVR")

plt.plot(test_scaled, y_test)
plt.plot(test_scaled, predictionPoly, color = 'blue', label='Polynomial model')
plt.plot(test_scaled, predictionLin, color = 'green', label='Linear model')
plt.plot(test_scaled, predictionrbf, color = 'red', label='RBF model')
plt.show()
```
